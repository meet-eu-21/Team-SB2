## Basic Imports

```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

## Create necessary folders -> /!\ Do not run twice the cell below

```python
#Create Folders -> /!\ Do not run twice : useless
cell_lines = ['GM12878', 'HMEC', 'HUVEC', 'IMR90', 'NHEK' ]
all_resol = ['100', '25']
path = 'Results/'
histo_path = os.path.join(path, 'Results - Histograms')
os.mkdir(histo_path)
path+= 'Results - Histograms/'
for cell in cell_lines:
    cell_path = os.path.join(path, cell)
    os.mkdir(cell_path)
    cell_path = path+cell+'/'
    for resol in all_resol:
        resol_path = os.path.join(cell_path, resol)
        os.mkdir(resol_path)
        
```

## Reading files

```python
cell_lines = ['GM12878', 'HMEC', 'HUVEC', 'IMR90', 'NHEK' ]
nb_chr = 23
data_100kb = np.empty((len(cell_lines),nb_chr),dtype=object)#[[[]]*nb_chr]*len(cell_lines)
data_25kb = np.empty((len(cell_lines),nb_chr),dtype=object)
cell_folders = [ f.path for f in os.scandir('Results/') if f.is_dir() ]
for cell_fold in cell_folders:
    if cell_fold=='Results/Results - Histograms':
        continue
    cell_line = cell_fold[8:]
    chr_folder = [ f.path for f in os.scandir('Results/'+cell_line+'/') if f.is_dir() ]
    index_Cell = cell_lines.index(cell_line)
    for chr_fold in chr_folder:
        chromo = chr_fold[8+len(cell_line)+1:]
        if chromo=='X':
            index = 22
        else:
            index = int(chromo)-1
        data_100 = 'Results/'+cell_line+'/'+chromo+"/100/Predictions.txt"
        data_25 = 'Results/'+cell_line+'/'+chromo+"/25/Predictions.txt"
        with open(data_100,'r') as f:
            lines = f.readlines()
            similarity_Leo_Contact = float(lines[1][31:-3])
            Best_cpt_Contact = int(lines[2][26:-1])
            similarity_Leo_Epi = float(lines[5][31:-3])
            Best_cpt_Epi = int(lines[6][26:-1])
            Consensus = float(lines[10][13:20])*100
            data_100kb[index_Cell][index] = [similarity_Leo_Contact, Best_cpt_Contact, similarity_Leo_Epi,
                                             Best_cpt_Epi, Consensus]

        """with open(data_25,'r') as f:
            lines = f.readlines()
            similarity_Leo_Contact = float(lines[1][31:-3])
            Best_cpt_Contact = int(lines[2][26:-1])
            similarity_Leo_Epi = float(lines[5][31:-3])
            Best_cpt_Epi = int(lines[6][26:-1])
            Consensus = float(lines[10][13:20])*100
            data_25kb[index_Cell][index] = [similarity_Leo_Contact, Best_cpt_Contact, similarity_Leo_Epi,
                                             Best_cpt_Epi, Consensus]"""

```

## Analyzing 100kb data

```python
#100kb
All_simi_Contact = np.zeros((len(cell_lines),nb_chr))
All_cpt_Contact = np.zeros((len(cell_lines),nb_chr))
All_simi_Epi = np.zeros((len(cell_lines),nb_chr))
All_cpt_Epi = np.zeros((len(cell_lines),nb_chr))
All_Consensus = np.zeros((len(cell_lines),nb_chr))

for cell in range(len(cell_lines)):
    for chromo in range(nb_chr):
        data = data_100kb[cell][chromo]
        if data==None:
            continue
        All_simi_Contact[cell][chromo] = data[0]
        All_cpt_Contact[cell][chromo] = data[1]
        All_simi_Epi[cell][chromo] = data[2]
        All_cpt_Epi[cell][chromo] = data[3]
        All_Consensus[cell][chromo] = data[4]
```

```python
#100kb
for cell in range(len(cell_lines)):
    cell_name = cell_lines[cell]
    path = 'Results/Results - Histograms/'+cell_name+'/100/'
    
    #similarity_Leo_Contact
    simi_contact = All_simi_Contact[cell]
    plt.figure(figsize=(16,9))
    locs, labels = plt.xticks()
    plt.xticks(np.arange(0, nb_chr+1, step=1))
    ax = plt.gca()  ;  ax.set_ylim([0, 100])
    plt.title(cell_name+" 100kb\nHMM Contact - Similarity with Leopold", size=20)
    plt.ylabel("Chromosome number", size=16)
    plt.ylabel("% of Similarity", size=16)
    for i in range(len(simi_contact)):
        plt.vlines(i+1, 0, simi_contact[i], linestyles ="solid", colors ="k")
    fname = path+'similarity_Leo_Contact.png'
    plt.savefig(fname)
    plt.close()
    
    #Best_cpt_Contact
    cpt_contact = All_cpt_Contact[cell]
    plt.figure(figsize=(16,9))
    locs, labels = plt.xticks()
    plt.xticks(np.arange(0, nb_chr+1, step=1))
    plt.title(cell_name+" 100kb\nHMM Contact - number of cpt predicted", size=20)
    plt.ylabel("Chromosome number", size=16)
    plt.ylabel("Number of cpt predicted", size=16)
    for i in range(len(cpt_contact)):
        plt.vlines(i+1, 0, cpt_contact[i], linestyles ="solid", colors ="k")
    fname = path+'Best_cpt_Contact.png'
    plt.savefig(fname)
    plt.close()
    
    #similarity_Leo_Epi
    simi_Epi = All_simi_Epi[cell]
    plt.figure(figsize=(16,9))
    locs, labels = plt.xticks()
    plt.xticks(np.arange(0, nb_chr+1, step=1))
    ax = plt.gca()  ;  ax.set_ylim([0, 100])
    plt.title(cell_name+" 100kb\nHMM Expr-Repr - Similarity with Leopold", size=20)
    plt.ylabel("Chromosome number", size=16)
    plt.ylabel("% of Similarity", size=16)
    for i in range(len(simi_Epi)):
        plt.vlines(i+1, 0, simi_Epi[i], linestyles ="solid", colors ="k")
    fname = path+'similarity_Leo_Expr-Repr.png'
    plt.savefig(fname)
    plt.close()
    
    #Best_cpt_Epi
    cpt_Epi = All_cpt_Epi[cell]
    plt.figure(figsize=(16,9))
    locs, labels = plt.xticks()
    plt.xticks(np.arange(0, nb_chr+1, step=1))
    plt.title(cell_name+" 100kb\nHMM Expr-Repr - number of cpt predicted", size=20)
    plt.ylabel("Chromosome number", size=16)
    plt.ylabel("Number of cpt predicted", size=16)
    for i in range(len(cpt_Epi)):
        plt.vlines(i+1, 0, cpt_Epi[i], linestyles ="solid", colors ="k")
    fname = path+'Best_cpt_Expr-Repr.png'
    plt.savefig(fname)
    plt.close()
    
    #Consensus
    simi_consensus = All_Consensus[cell]
    plt.figure(figsize=(16,9))
    locs, labels = plt.xticks()
    plt.xticks(np.arange(0, nb_chr+1, step=1))
    ax = plt.gca()  ;  ax.set_ylim([0, 100])
    plt.title(cell_name+" 100kb\nConsensus Similarity", size=20)
    plt.ylabel("Chromosome number", size=16)
    plt.ylabel("% of Similarity", size=16)
    for i in range(len(simi_consensus)):
        plt.vlines(i+1, 0, simi_consensus[i], linestyles ="solid", colors ="k")
    fname = path+'Consensus.png'
    plt.savefig(fname)
    plt.close()
```
